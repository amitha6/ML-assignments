```python
def sigmoid(x):
    output = 1/(1+np.exp(-x))
    return output

def sigmoid_derivative(output):
    return output*(1-output)

def relu(x):
    return x
```

```python
def softmax(Z):
    A = np.exp(Z) / sum(np.exp(Z))
    return A
```

```python
def relu(Z):
    return np.maximum(Z, 0)
```

```python
def linear(Z):
    return Z
```
